## Outlier detection

```python
# Use plotting to identify outliers in the dataset

for i in df:
        r = df[i].dtypes
        if r == 'float64':
            plt.scatter(range(len(df[i])), df[i], label=i)
            plt.legend()
            plt.show()
```

```python
df = df[(df.meas_length < 40)]
df = df[(df.meas_width < 30)]
df = df[(df.meas_depth < 30)]
```

```python
# Now lets check if measured parameters are 0 or not
df[(df.meas_width == 0) | (df.meas_depth == 0) | (df.meas_length == 0) | (df.depth_percent == 0) | (df.table_percent == 0)]
```

```python
#since it is not substantial we can remove this and store in a different DF to check for model accuracy
df2 = df.copy()
df2 = df2[(df2.meas_width != 0)]
df2 = df2[(df2.meas_depth != 0)] 
df2 = df2[(df2.meas_length != 0)]
df2 = df2[(df2.depth_percent != 0) ]
df2 = df2[(df2.table_percent != 0)]
```

```python
print(df.shape)
print(df2.shape)
```

## Step 3: Feature Selection & Data splitting

```python
# Let's perform quantitative analysis by feature correlation study

from sklearn.preprocessing import StandardScaler, MinMaxScaler
ss = MinMaxScaler()

corr = df2.corr()
sns.heatmap(corr)
```

```python
# First define all the features
features = []
for i in df2:
        if i != 'total_sales_price':
            features.append(i)
print(features)
```

```python
X = df[features]
y = df2['total_sales_price']

print(X.shape)
print(y.shape)
'''
#miniX = np.min(X)
#maxiX = np.max(X)
miniy = np.min(y)
maxiy = np.max(y)
#print(miniX)
print(maxiX)
#X = (X-miniX)/(maxiX - miniX)
y = (y-miniy)/(maxiy - miniy)
'''
#X = ss.fit_transform(X)
#y = ss.fit_transform(y.values.reshape(-1,1))
print(X.shape)
print(y.shape)
```

```python
#using f_regression score we can select k best features
from sklearn.feature_selection import SelectKBest as skb
from sklearn.feature_selection import f_regression

# Ensure X and y have consistent lengths
# Assuming X and y are defined elsewhere in the notebook, you might need to preprocess or trim them so they match.
# This is a placeholder step. You'll need to adjust it based on your specific situation.
# For example, if X has more rows than y, you might need to trim X to match y's length, or vice versa.
# Here's a simple example of trimming X if it has more rows than y. Adjust according to your data's needs.

# Ensure X and y have the same number of samples
# This is a critical step before fitting the model
if len(X) > len(y):
    X = X[:len(y)]
elif len(y) > len(X):
    y = y[:len(X)]

#first we analyze all scores
fs = skb(f_regression, k='all')
fs.fit(X, y)
for i in range(len(fs.scores_)):
    print('Feature %d: %f' % (i, fs.scores_[i]))
```

```python
# Now we will select 15 Best Features out of 24 and get their names

fs = skb(f_regression, k=15)
Xnew = fs.fit_transform(X,y)
f = np.array(features)
filt = fs.get_support()
f = f[filt]
print(f, len(f))
```

```python
Xnew = X
y = y.values.reshape(-1,1)
print(Xnew.shape)
print(y.shape)
```

```python
# Perform final check
print(Xnew, Xnew.shape)
print(X,y.shape)
```

```python
# Splitting the dataset into training and testing 
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(Xnew, y, test_size= 0.2, random_state=40)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## Step 4: Modelling regression using Linear Regression 

```python
from sklearn.linear_model import LinearRegression 

mlr = LinearRegression()

mlr.fit(X_train, y_train)

print('intercept:', mlr.intercept_)
print('Coeff: ', list(zip(X, mlr.coef_)))
```

```python
y_pred_mlr = mlr.predict(X_test)
```

```python
# Calculate R2 and Mean Absolute error

from sklearn import metrics

MAE = metrics.mean_absolute_error(y_test, y_pred_mlr)
r2 = mlr.score(X_test, y_test)*100

print('R squared score: ', r2)
print('Mean Absolute Error: ', MAE)
```

## Step 5: Modelling regression using Artificial Neural Network (ANN)

```python
import tensorflow
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Define the model architecture
model = Sequential()
model.add(Dense(units=100, input_dim=24, kernel_initializer='normal', activation='relu'))
model.add(Dense(units=50, kernel_initializer = 'normal', activation= 'relu'))
model.add(Dense(1, kernel_initializer = 'normal'))

# Since this is a curve fitting problem loss is MSE
# So lets add a history to track the loss
model.compile(loss='mean_squared_error', optimizer='adam')
history = model.fit(X_train, y_train, validation_split = 0.2, batch_size= 2000, epochs=50, verbose=1 )
```

```python
# Plot loss curve
'''
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
'''
pd.DataFrame(history.history).plot(figsize =(10,10))
plt.show()
```

From the above we can see that there is a steady smooth decrease in training and validation loss. so our model is not overfitting.

```python
# Perform Predictions

pred = model.predict(X_test)

from sklearn.metrics import r2_score
r2 = r2_score(y_test, pred)*100

print('R Squared: ', r2)
```

We are getting R2 score of 34% which is not very high. We can imporve this fyrther by choosing fewer features and doing some more pre processing.

```python
#plotting the fit

print(pred[5024])
print(y_test[5024])
plt.figure(figsize=(50,10))
plt.plot(y_test[5000:5100],'ro--',label='actual')
plt.plot(pred[5000:5100], 'b+-', label='predicted')
plt.legend()
plt.show()
```

From the above we can see a descent fit between predicted values and actual values
