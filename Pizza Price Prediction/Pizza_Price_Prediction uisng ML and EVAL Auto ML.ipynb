## Step 3: Feature Engineering

#### One Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder
import category_encoders as ce
def one_hot_encoding(df,col):
    one_hot_encoder=ce.OneHotEncoder(cols=col,return_df=True,use_cat_names=True)
    df_final = one_hot_encoder.fit_transform(df)
    return df_final
def one_hot(df, column):
    df = one_hot_encoding(df,column)
    return df
```

```python
df = one_hot(df, 'company')
df = one_hot(df, 'topping')
df = one_hot(df, 'variant')
df = one_hot(df, 'size')
df = one_hot(df, 'extra_sauce')
df = one_hot(df, 'extra_cheese')
```

```python
df = one_hot(df, 'extra_mushrooms')
```

```python
df.head()
```

```python
df.dtypes
```

```python
x = df.drop("price",axis = 1)
y = df["price"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2)
```

High Predictive Power: With an RÂ² score of 0.94708, your model explains about 94.7% of the variance, indicating it is quite effective in capturing the overall trend in the data.

Significant Prediction Errors: Despite the high RÂ², the MAE (4740.21965) and RMSE (11509.70057) suggest that the model's individual predictions can vary considerably from the actual values, indicating a notable degree of error.

Possible Impact of Outliers: The larger RMSE compared to MAE suggests the presence of outliers or a few large errors, which may be skewing the predictions.

Model Suitability Depends on Application: The model is good for understanding general trends but may be less reliable for precise predictions for each data point.

Potential for Improvement: There may be a need to investigate and potentially improve the model, possibly by addressing overfitting, re-evaluating feature selection, or exploring different modeling techniques.

### We will use the same train and test dataset for this

```python
import evalml
```

```python
evalml.problem_types.ProblemTypes.all_problem_types
```

```python
X_train, X_test, y_train, y_test = evalml.preprocessing.split_data(x, y, problem_type='regression')
```

**Running the Auto ML to select best Algorithm**

```python
from evalml.automl import AutoMLSearch
automl = AutoMLSearch(X_train=X_train, y_train=y_train, problem_type='regression')
automl.search()
```

```python
automl.rankings
```

```python
automl.best_pipeline
```

```python
best_pipeline=automl.best_pipeline
```

```python
automl.describe_pipeline(automl.rankings.iloc[0]["id"])
```

```python
best_pipeline.score(X_test, y_test, objectives=["R2"])
```

The provided metric is an RÂ² score of 0.968590569595417. Here's a concise conclusion based on this score:

Highly Effective Model: An RÂ² score of approximately 0.969 indicates that the model is highly effective in explaining the variance in the dependent variable. This suggests that about 96.9% of the variance in the target variable can be predicted from the independent variables.

Strong Predictive Power: Such a high RÂ² value typically reflects a strong predictive power of the model, assuming that the RÂ² score is not a result of overfitting.

However, while the RÂ² score is an important indicator of model performance, it should not be the sole metric for evaluating a model's effectiveness. It's also essential to consider other metrics like MAE, RMSE, and to check for overfitting, to get a comprehensive view of the model's performance.

### Making a model based on R2 objective

```python
automl_r2 = AutoMLSearch(X_train=X_train, y_train=y_train,
                          problem_type='regression',
                          objective='R2',
                          
                          max_batches=1,
                          optimize_thresholds=True)

automl_r2.search()
```

```python
automl_r2.rankings
```

From the above, it seems that Random forest regressor is the best performing model with high R2 score theteby keeping high accuracy levels

```python
automl_r2.describe_pipeline(automl_r2.rankings.iloc[0]["id"])
```

```python
best_pipeline_r2 = automl_r2.best_pipeline
```

```python
best_pipeline_r2.score(X_test, Y_test,  objectives=["R2"])
```
