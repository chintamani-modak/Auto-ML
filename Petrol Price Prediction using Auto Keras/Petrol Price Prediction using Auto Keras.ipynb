## Step 2: Data Analysis and Split the data into Train & Test 

```python
df1.isnull().sum()
```

Now, as you can see from the above, this is a scatter plot.This is the wave which shows that how our plot is, how our prices are distributed.These are some of the outliers in the 180 part.We can drop this for sure because these are of no importance or will not make much difference.

```python
df3 = df2[df2.Petrol<160]
plt.scatter(df3.Date, df3.Petrol, color='red', label='Prices', s=50)
```

```python
# Reset the index to convert this data into a more convenient form of two into to try to give my data into a minmaxscaler.

df4=df3.reset_index()['Petrol']
df4.head()
```

```python
# Use MinMax Scaler to transform the values
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
df4=scaler.fit_transform(np.array(df4).reshape(-1,1))
```

```python
df4
```

```python
training_size=int(len(df4)*0.65)
test_size=len(df4)-training_size
train_data,test_data=df4[0:training_size,:],df4[training_size:len(df4),:1]
```

```python
train_data.shape
test_data.shape
```

## Step 3: Data Preprocessing 

```python
import numpy as np
# Convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return np.array(dataX), np.array(dataY)
```

```python
# reshape into X=t, t+1, t_2, t+3 and t+4
time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)
```

```python
# Import necessary libraries
import tensorflow as tf
import autokeras as ak
```

```python
reg = ak.StructuredDataRegressor(overwrite=True, max_trials=4)
```

```python
reg.fit(X_train, y=y_train, verbose=0)
```

```python
# Evaluate the model
mae,_=reg.evaluate(X_test, y_test, verbose=0)

# Use the model to make a prediction
yhat_test = reg.predict(X_test)

# Now get the best performing model
model = reg.export_model()
```

```python
# Summarise the loaded model
model.summary()
```

```python
yhat_train = reg.predict(X_train)
```

```python
train_predict = scaler.inverse_transform(yhat_train)
test_predict = scaler.inverse_transform(yhat_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,yhat_test)
```

As we can see from the above, Using AutoKeras, we are getting a MSE of 0.06 on test data and MSE of 0.03 which is indicating that  our model is predicted satisfactorarily. The magnitude of the difference in MSE (0.03 vs. 0.06) is not extremely large, but it's significant enough to warrant investigation and potential adjustments to your model or training process.

In summary, while a certain degree of difference in performance between training and test sets is normal, a consistent and notable discrepancy like the one we're observing is typically a sign of overfitting, and it suggests that your model may not be generalizing well to new data. Suggested alternative solution would be to use **ARIMA** to compare the models evaluation between different models to finally decide on which model to choose for Production and implementation.

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from matplotlib import pyplot as plt
from pandas import read_csv
from pandas import datetime 
from matplotlib import pyplot

from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
df = pd.read_csv("train_data.csv")
df.head()
```

```python
df.rename(columns={"Petrol (USD)": "Petrol"}, inplace = True)
df.head()
```

```python
df2 = df.dropna()
df3=df2[df2.Petrol<160]
df3.head()
```

```python
df4=df3.reset_index()['Petrol']
df4.head()
```

```python
model1 = ARIMA(df4.values, order=(5,1,0))
model_fit1 = model1.fit()
output1= model_fit1.forecast(steps=30)
output1
```
